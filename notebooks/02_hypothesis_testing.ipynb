```python
import pandas as pd
import numpy as np

from src.data_loader import load_data
from src.hypothesis_tests import perform_ttest
```

```python
df = load_data("../data/machineleariningRating_v3.txt")
```

```python
df["LossRatio"] = (
    df["TotalClaims"] /
    df["TotalPremium"]
)
```

```python
df["LossRatio"] = np.where(
    df["TotalPremium"] > 0,
    df["TotalClaims"] / df["TotalPremium"],
    0
)
```

```python
df["Province"].value_counts()
```

```python
gauteng = df[
    df["Province"] == "Gauteng"
]["LossRatio"]

western_cape = df[
    df["Province"] == "Western Cape"
]["LossRatio"]
```

```python
stat, p_value = perform_ttest(
    gauteng,
    western_cape
)

print("T-statistic:", stat)
print("P-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject H0")
    print("There is a statistically significant risk difference.")
else:
    print("Fail to Reject H0")
    print("No statistically significant difference detected.")
```

The hypothesis test indicates a statistically significant
difference in loss ratio between Gauteng and Western Cape
(p < 0.05).

This suggests regional risk differs materially between
these provinces. ACIS may consider province-specific
pricing adjustments or targeted underwriting strategies.

```python
results = pd.DataFrame({
    "Hypothesis": ["Province Risk Difference"],
    "Test": ["T-Test"],
    "P-Value": [p_value],
    "Decision": [
        "Reject H0" if p_value < 0.05
        else "Fail to Reject H0"
    ]
})

results
```
